# ***KMeans***

```python
clustering1 = KMeans(n_clusters = 3)
```

```python
clustering1.fit(df[['Annual Income (k$)']])
```

```python
clustering1.labels_
```

```python
df['Income Clustering'] = clustering1.labels_
df.head()
```

```python
df["Income Clustering"].value_counts()
```

```python
clustering1.inertia_
```

# *Elbow method*

```python
intertia_scores = []
for i in range(1,11):
  kmean = KMeans(n_clusters = i)
  kmean.fit(df[['Annual Income (k$)']])
  intertia_scores.append(kmean.inertia_)
```

```python
intertia_scores
```

```python
plt.plot(range(1,11),intertia_scores)
```

# **Bivariate Clustering**

```python
clustering2 = KMeans(n_clusters = 5)
clustering2.fit(df[['Annual Income (k$)','Spending Score (1-100)']])
clustering2.labels_

df['Spending_and_Income_Cluser'] = clustering2.labels_

df.head()
```

```python
intertia_scores2 = []
for i in range(1,11):
  kmean2 = KMeans(n_clusters = i)
  kmean2.fit(df[['Annual Income (k$)','Spending Score (1-100)']])
  intertia_scores2.append(kmean2.inertia_)
plt.plot(range(1,11),intertia_scores2)
```

```python
df.columns
```

```python
center = pd.DataFrame(clustering2.cluster_centers_)
center.columns = ["x","y"]
center
```

```python
plt.figure(figsize = (10,8))
#show centroid point
plt.scatter(x = center["x"],
            y = center["y"],
            c = "black" , #color
            marker = '*', #เปลี่ยนให้เป็น สัญลักณ์อะไร
            s = 100)


sns.scatterplot(data= df, x = 'Annual Income (k$)',
                y = 'Spending Score (1-100)',
                hue = "Spending_and_Income_Cluser",
                palette = 'tab10'
                )
```

```python
pd.crosstab(df['Spending_and_Income_Cluser'],df['Gender'])
```

# **Multivariate clustering**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
```

```python
df.head()
```

```python
#one hot coding
dff = pd.get_dummies(df,drop_first = True)
dff
```

```python
dff.columns
```

```python
dff = dff[['Age', 'Annual Income (k$)', 'Spending Score (1-100)', 'Gender_Male']]

dff.head()
```

```python
dff = scale.fit_transform(dff)

dff = pd.DataFrame(scale.fit_transform(dff))
dff.head()
```

```python
intertia_scores3 = []
for i in range(1,11):
  kmean3 = KMeans(n_clusters = i)
  kmean3.fit(dff)
  intertia_scores3.append(kmean3.inertia_)
plt.plot(range(1,11),intertia_scores3)
```
